```python
# Encoding the Categorical Data
```

```python
data = pd.get_dummies(data = df, columns = ['job', 'marital', 'education', 'contact',
       'month', 'poutcome'])
```

```python
data
```

```python
# To see all columns
pd.set_option("display.max.columns", None)
```

```python
data
```

```python
data['default'] = data['default'].map({'yes' : 1, 'no' : 0})
```

```python
data
```

```python
def binary(data,colname) :
    data[colname] = data[colname].map({'yes' : 1, 'no' : 0})
    return data[colname]
```

```python
binary(data, 'housing')
```

```python
binary(data,'loan')
```

```python
binary(data,'y')
```

```python
data
```

```python
# Shifting y to the first position of the dataset
temp_col = data.pop('y')
data.insert(0, 'y', temp_col)
data
```

```python
# Define x and y
```

```python
x = data.iloc[:,1:]
x
```

```python
y = data['y']
y
```

```python
#Model Building
```

```python
# Spliting the data into Train and Test
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size= 0.2, random_state = 42)
```

```python
xtrain
```

```python
ytrain
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg = LogisticRegression()
```

```python
lg.fit(xtrain, ytrain)
```

```python
#lg.coef_
```

```python
ypred = lg.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(ytest,ypred)
```

```python
sns.heatmap(confusion_matrix(ytest,ypred), annot = True, fmt = 'g')
```

```python
print(classification_report(ytest, ypred))
```

```python
print(ytest[:30].values)
print(ypred[:30])
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest, ypred)
```

```python
# As we can observe that are model is performing well
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
proba = lg.predict_proba(xtest)[:,1]
```

```python
proba
```

```python
auc = roc_auc_score(ytest,proba)
auc
```

```python
fpr, tpr, threshold = roc_curve(ytest, proba)
```

```python
plt.plot(fpr,tpr, label = 'AUC : %0.2f'%auc)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
```
