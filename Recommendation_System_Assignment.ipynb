```python
# User to User Based
```

```python
pivot = pd.pivot_table(df, index='User.ID', columns='Book.Title', values=['Book.Rating'] , aggfunc='count')
pivot
```

```python
# Filling up NA values
```

```python
pivot.fillna(0, inplace = True)
```

```python
pivot
```

```python
# Calculate the Distances and Similarities
```

```python
from sklearn.metrics import pairwise_distances
```

```python
pd.DataFrame(pairwise_distances(pivot,metric = 'cosine'))
```

```python
#Similarity Calculation
```

```python
similarity = pd.DataFrame(1-pairwise_distances(pivot,metric = 'cosine'))
similarity
```

```python
#Convert every single diagonal values to 0
# Change Default index Number on both row and column side
```

```python
np.fill_diagonal(similarity.values,0)
```

```python
similarity
```

```python
similarity.index = df['User.ID'].unique()
similarity.columns = df['User.ID'].unique()
similarity
```

```python
# Find Similar Users
```

```python
similarity.idxmax()
```
