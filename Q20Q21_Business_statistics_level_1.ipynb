```python
from scipy import stats
from scipy.stats import norm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.dpi'] = 150
plt.rcParams['figure.figsize'] = (10,5)
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Users\ahmed\Downloads\Cars.csv")
df
```

```python
df.describe()
```

```python
mean = df['MPG'].mean()
mean
```

```python
std = df['MPG'].std()
std
```

```python
z_scores = stats.zscore(df['MPG'])
z_scores
```

```python
mpg = 38
z=(mpg-mean)/std
z
```

```python
p_value = norm.cdf(z)
p_value
```

```python
#For mpg > 38
p=1-p_value
p
```

```python
sns.distplot(z_scores, hist = False)
plt.axvline(x = z, linestyle = '--', color = 'red')
```

```python
stats.norm.cdf(38,mean,std)
```

```python
#For P(MPG>38)
1-stats.norm.cdf(38,mean,std)
```

```python
#For P(MPG<40)
stats.norm.cdf(40,mean,std)
```

```python
#For P(20<MPG<50)
stats.norm.cdf(50,mean,std)-stats.norm.cdf(20,mean,std)
```
